# Escalado
Conjunto de datos: iris (incluido en sklearn)
Este conjunto de datos es compacto y perfecto para demostrar técnicas de escalado.

```python
# Importar librerías necesarias
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar el conjunto de datos
iris = load_iris(as_frame=True)
data = iris.data
```

```python
# Visualizar las estadísticas iniciales
print("Estadísticas iniciales:")
data.describe()
```

```python
# Aplicar MinMaxScaler
minmax_scaler = MinMaxScaler()
data_minmax = pd.DataFrame(minmax_scaler.fit_transform(data), columns=data.columns)
```

```python
# Aplicar StandardScaler
standard_scaler = StandardScaler()
data_standard = pd.DataFrame(standard_scaler.fit_transform(data), columns=data.columns)
```

```python


print("Estadísticas iniciales:")
print(data.describe())


# Escalado MinMax

minmax_scaler = MinMaxScaler()
data_minmax = pd.DataFrame(minmax_scaler.fit_transform(data), columns=data.columns)

print("\nEstadísticas con MinMaxScaler:")
print(data_minmax.describe())


# Escalado StandardScaler

standard_scaler = StandardScaler()
data_standard = pd.DataFrame(standard_scaler.fit_transform(data), columns=data.columns)

print("\nEstadísticas con StandardScaler:")
print(data_standard.describe())

data.describe()
data_minmax.describe()

print("A diferencia de StandardScaler, MinMax solo mueve los datos entre 0 y 1.")





```

### 2. Crea una gráfica de la distribución de la variable 'sepal length (cm)' antes y después del escalado.

```python
# 2. Histograma antes - despues del MinMax '


plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(data['sepal length (cm)'], bins=20)
plt.title("Antes del escalado")
plt.xlabel("sepal length (cm)")
plt.ylabel("Frecuencia")

plt.subplot(1,2,2)
plt.hist(data_minmax['sepal length (cm)'], bins=20)
plt.title("Después del escalado (MinMax)")
plt.xlabel("sepal length (cm)")

plt.tight_layout()
plt.show()
```

### 3. Crea una gráfica de dispersión entre 'sepal length (cm)' y 'sepal width (cm)' con los datos escalados por StandardScaler.

```python
# 3. Gráfica de dispersión con datos escalados (StandardScaler)

plt.figure(figsize=(7,5))
plt.scatter(
    data_standard['sepal length (cm)'],
    data_standard['sepal width (cm)']
)
plt.title("Dispersión estandarizada (StandardScaler)")
plt.xlabel("sepal length (cm) - escalado")
plt.ylabel("sepal width (cm) - escalado")
plt.grid(True)
plt.show()
```
